## LLM
using HuggingFace opensource model .

```python
import os 
from dotenv import load_dotenv
from langchain_huggingface import HuggingFaceEndpoint , ChatHuggingFace
```

```python
#load environment 
load_dotenv()
```

```python
#check token 
token = os.getenv("HUGGINGFACEHUB_API_TOKEN")
if not token:
    print("Error: HUGGINGFACEHUB_API_TOKEN is not set in the .env file.")
else:
    print("Done")
```

```python
#  Setup the Endpoint
repo_id = "meta-llama/Llama-3.1-8B-Instruct"

llm_H = HuggingFaceEndpoint(
    repo_id=repo_id,
    task="text-generation",
    max_new_tokens=512,
    temperature=0.7,
    huggingfacehub_api_token=token,
)
```

```python
#  Wrap it in ChatHuggingFace (This handles the "conversational" format for you)
llm = ChatHuggingFace(llm=llm_H)
```
